```python
# let's break data into training and test data.
from sklearn.model_selection import train_test_split

x_train_data,x_test_data,y_train_label,y_test_label = train_test_split(x,y,test_size=0.1,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

y_train_label_log = np.log(y_train_label)
ninja = StandardScaler()
y_train_label_reshape = y_train_label_log.values.reshape(-1,1)

y_train_label_reshape_scaled = ninja.fit_transform(y_train_label_reshape).flatten()
y_train_label_reshape_scaled
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
import joblib
```

```python
num_features = ['Area','BHK','Bathroom']
cat_features = ['Area_Name','Transaction','Furnishing','Type']

# Define the numerical and categorical preprocessing
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_features),
        ('cat', OneHotEncoder(), cat_features)
    ]
)
```

```python
# Define the full pipeline with preprocessing and the model
full_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor)
])
```

```python
processed_input_data = full_pipeline.fit_transform(x_train_data)
```

```python
reg_model = LinearRegression()
```

```python
reg_model.fit(processed_input_data,y_train_label_reshape_scaled)
```

```python
x_test_data.head()
```

```python
processed_test_data = full_pipeline.transform(x_test_data)
```

```python
predictions = reg_model.predict(processed_test_data)
```

```python
predictions_1 = predictions.flatten()
predictions_2 = predictions_1.reshape(-1,1)
predictions_3 = ninja.inverse_transform(predictions_2)
predictions_4 = np.exp(predictions_3)
```

```python
predictions_4[:5]
```

```python
y_test_label.head(5)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test_label,predictions_4)
```

```python
# Save the full pipeline
joblib.dump(full_pipeline, 'full_pipeline.pkl')

#Save linear regression model
joblib.dump(reg_model,'reg_model.pkl')

# Save the scaler
joblib.dump(ninja, 'ninja.pkl')
```
